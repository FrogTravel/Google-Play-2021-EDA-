### What is the proportion of free to paid apps in each category? What is the category with the biggest relative percentage amount of paid apps?
Draw bar plot stacked vertically 

```python
number_of_free_apps_per_category = data[~data["Free"]].groupby("Category").size().reset_index()
number_of_total_apps_per_category = data.groupby("Category").size().reset_index()

result = pd.merge(left= number_of_total_apps_per_category, right = number_of_free_apps_per_category, on = "Category").rename(columns={"0_x": "Total", "0_y": "Paid"})
result["Proportion"] = result["Paid"] / result["Total"]
result = result.sort_values(by="Proportion", ascending=False)
result.plot(kind="bar", stacked=True, x = "Category", y="Proportion", figsize=(9, 4), title="Proportion of Paid to Free apps")

result
```

The highest proportion of paid apps are in "Role Playing" category with 7% of Paid apps

### Paid vs free apps — rating comparison

```python
avg_rating_per_category_per_paid = data.groupby(["Category", "Free"])["Rating"].mean().unstack().sort_values(True, ascending=False).rename(columns={True: "Free", False: "Paid"})   # sort by Free apps rating

avg_rating_per_category_per_paid.plot(kind="bar", figsize=(9, 4), title="Average rating per category for paid and free apps")
```

```python
import matplotlib.pyplot as plt

# Claude 

df = data.groupby(["Category", "Free"])["Rating"].mean().unstack()

fig, ax = plt.subplots(figsize=(8, 8))
ax.scatter(df[True], df[False], alpha=0.7)

for cat in df.index:
    ax.annotate(cat, (df.loc[cat, True], df.loc[cat, False]), fontsize=9)

# Diagonal = equal ratings
lims = [df.min().min(), df.max().max()]
ax.plot(lims, lims, "r--", alpha=0.5)

ax.set_xlabel("Free Rating")
ax.set_ylabel("Paid Rating")
ax.set_title("Free vs Paid avg rating by category")
```

```python
## Claude, fixed by me

df = data.groupby(["Category", "Free"])["Rating"].mean().unstack()
diff = (df[True] - df[False]).sort_values()
ax = diff.plot(kind="barh", figsize=(7, 12), title="Free minus Paid avg rating by category")

plt.axvline(0, color="black", linewidth=0.8)
plt.xlabel("Rating difference (Free − Paid)")

# Annotate sides
ax.text(diff.min() * 0.6, len(diff) * 0.95, "← Paid rated higher", fontsize=10, color="gray")
ax.text(diff.max() * 0.2, 0.6, "Free rated higher →", fontsize=10, color="gray")

```

Libraries and Demo are rated higher when Paid

# "Hidden gems" analysis

### Which apps have a high rating but relatively low install count (underrated apps)?

```python
def safe_divide(a, b):
    if b == 0:
        return 0
    else:
        return a / b

# How many ratings people leave per install? (as a proxy for user engagement)
data["prop"] = data.apply(lambda row: safe_divide(row["Rating Count"], row["Maximum Installs"]), axis=1)

data.sort_values(by="prop", ascending=False).head(3)
```

Top 3 are such hidden gems as Radar Detector with crazy package name, All in one Status Saver (I have no idea what is it) and Chopper Slice - Master Chef! None of this app is available in Google Play
